```python
import networkx as nx
import matplotlib.pyplot as plt
G = nx.Graph()
# add nodes from list
G.add_nodes_from([1,2,3,4,5,6])
# add edges from list
G.add_edges_from([(1,2),(1,3),(1,4),(1,6),(2,3),(2,4),(3,6)])

# draw the graph and the nodes that do not have any edges will not be shown
#nx.draw(G, with_labels=True, font_weight='bold')

diG = nx.DiGraph()
# add nodes from list
diG.add_nodes_from([1,2,3,4,5,6])
# add edges from list
diG.add_edges_from([(1,2),(2,3),(2,4),(3,2),(3,1),(4,1),(6,1),(6,3)])

# draw the graph and the nodes that do not have any edges will not be shown
#nx.draw(diG, with_labels=True, font_weight='bold')

```

```python
#1 The corresponding adjacency matrices.
adj_G = nx.adjacency_matrix(G).todense()
adj_diG = nx.adjacency_matrix(diG).todense()
#print(adj_G)
#print(adj_diG)

#2 The corresponding link lists.
link_G = [x for x in nx.generate_edgelist(G)]
link_diG = [x for x in nx.generate_edgelist(diG)]
#print(link_G)
#print(link_diG)

#3 In the (a) network, how many paths (with possible repetition of nodes and links) of length 3 exist starting from node 1 and ending at node 3?
all_paths_G = [path for path in nx.all_simple_paths(G, source=1, target=3, cutoff=3)]
#print("Total paths (1->3): {}".format(all_paths))
# In the (b) network, how many paths (with possible repetition of nodes and links) of length 3 exist starting from node 1 and ending at node 3?
all_paths_diG = [path for path in nx.all_simple_paths(diG, source=1, target=3, cutoff=3)]
#print("Total paths (1->3): {}".format(all_paths))

# 4 count the number of cycles of length 4 in both networks
# In the (a) network, how many cycles of length 4 exist?
all_cycles_G = [cycle for cycle in nx.simple_cycles(G,4)]
#print("Total cycles of length 4: {}".format(all_cycles_G))
# In the (b) network, how many cycles of length 4 exist?
all_cycles_diG = [cycle for cycle in nx.simple_cycles(diG,4)]
#print("Total cycles of length 4: {}".format(all_cycles_diG))
```

```python
# bipartite graph
biG = nx.Graph()
blue_nodes = [1,2,3,4,5,6]
green_nodes = [7,8,9,10,11]
# add nodes from list
biG.add_nodes_from(blue_nodes, bipartite=0)
biG.add_nodes_from(green_nodes, bipartite=1)
# add edges from list
biG.add_edges_from([(1,7),(2,9),(3,7),(3,8),(3,9),(4,9),(4,10),(5,9),(5,11),(6,11)])
# bipartite graph layout
pos_b = nx.bipartite_layout(biG, blue_nodes)
# Rotate the layout 90 degrees by swapping x and y coordinates
rotated_pos = {node: (y, -x) for node, (x, y) in pos_b.items()}

nx.draw_networkx_nodes(biG, rotated_pos, nodelist=blue_nodes, node_color='b')
nx.draw_networkx_nodes(biG, rotated_pos, nodelist=green_nodes, node_color='g')
nx.draw_networkx_edges(biG, rotated_pos, width=1.0, alpha=0.5)
nx.draw_networkx_labels(biG, rotated_pos, font_size=10, font_family='sans-serif')

plt.axis("off")  # Turn off axis labels and ticks
# Display the rotated graph
plt.show()
```

```python
#1 Construct its adjacency matrix. 
adj_biG = nx.bipartite.biadjacency_matrix(biG, row_order=blue_nodes, column_order=green_nodes).todense()
# print(adj_biG)

#2 get the projection of the bipartite graph onto the blue nodes
biG_blue = nx.bipartite.projected_graph(biG, blue_nodes)
# get the adjacency matrix of the projection
adj_biG_blue = nx.adjacency_matrix(biG_blue).todense()
#print(adj_biG_blue)
#v visualise the projection
#nx.draw(biG_blue, with_labels=True, font_weight='bold')

# get the projection of the bipartite graph onto the green nodes
biG_green = nx.bipartite.projected_graph(biG, green_nodes)
# get the adjacency matrix of the projection
adj_biG_green = nx.adjacency_matrix(biG_green).todense()
#print(adj_biG_green)
# visualise the projection
#nx.draw(biG_green, with_labels=True, font_weight='bold')

#3 Calculate the average degree of the purple nodes and the average degree of the green nodes in the bipartite network
blue_degrees = biG.degree(blue_nodes)
avg_blue = sum([t[1] for t in blue_degrees])/len(blue_degrees)
print("Avg degree of blue from graph: {}".format(avg_blue))
green_degrees = biG.degree(green_nodes)
avg_green = sum([t[1] for t in green_degrees])/len(green_degrees)
print("Avg degree of green from graph: {}".format(avg_green))

# projections degrees
blue_degrees_proj = biG_blue.degree()
avg_blue_proj = sum([t[1] for t in blue_degrees_proj])/len(blue_degrees_proj)
print("Avg degree of blue from projection: {}".format(avg_blue_proj))
green_degrees_proj = biG_green.degree()
avg_green_proj = sum([t[1] for t in green_degrees_proj])/len(green_degrees_proj)
print("Avg degree of green from projection: {}".format(avg_green_proj))
```

# Part 3: Plotting degree distributions

As always we'll learn about degree-distribution plotting by creating a `notebook` and trying it out. 

As a little public service announcement about this exercise: I know that you've probably tried binning data before. And that there are many functions designed to make binning easy, so there is a good chance that you think that binning data into bins is something that you don't have to worry about.

**BUT** ... with this exercise, you may find that binning isn't always trivial. And that there are key things to think about every time you need to put data into buckets. So please read the exercise carefully. And follow my instructions carefully.


> 
> *Exercises*:
> 
> Begin by importing the right packages. Start by importing `matplotlib.pyplot` (for plotting), `numpy` (for binning and other stuff), 
> `random` (for generating random numbers), and `networkx` (for generating networks.)
> 
> * Binning real numbers
>   * Let's do a gentle start and use the `random` library generate 5000 data points from a Gaussian distribution with $\mu = 2$ and $\sigma = 0.125$.
>   * Now, let's use `numpy.histogram` to bin those number into 10 bins. What does the `numpy.histogram` function return? Do the two arrays have the same length?
>   * Then we use `matplotlib.pyplot.plot` to plot the binned data. You will have to deal with the fact that the counts- and bin-arrays have different lengths. Explain how you deal with this problem and why.
> * Binning integers
>   * But binning real numbers into a fixed number of bins is easy when `numpy.histogram` does all the work and finds the right bin boundaries for you. 
> Now we'll generate a bunch of integers and set the bin boundaries manually. This time, let's grab data from a Poisson distribution. As it turns out 
> `numpy` also has some convenient random number generators. Use `numpy.random.poisson` to generate 5000 numbers drawn from a Poisson distribution 
> characterized by $\lambda = 10$. Find the maximum and minimum value of your 5000 random numbers. 
>   * Instead of simply specifying the number of bins for `numpy.histogram`, let's specify the bins we want using a vector. 
> Create a vector $v$
> that results in a binning that puts each integer value in its own bin and where the first bin contains the minimum number you found above, 
> and the last bin contains the maximum number. 
> Use the vector by setting `numpy.histogram`'s `bin` parameter as `bin = ` $v$. 
> What is the sum over bin counts? Explain how the binning-vectors first and last element 
> relates to the min and max from the Poisson distribution.
>   * Now, use a bar chart (`matplotlib.pyplot.bar`) to plot the distribution 
> * Binning and plotting degree distributions. 
>   * Let's generate the Erdös-Renyi (ER) network which has a degree distribution that matches the Poisson distribution above. 
> First we have to figure out which values the ER parameters (_N_ and _p_) should assume. It's easy to see that $N = 5000$, but how do you find $p$? 
> **Hint**: The parameter $\lambda$ in the Poisson distribution corresponds to the average degree, so you have to find a $p$ that results in an average 
> degree, $\langle k \rangle = 10$. And you know that $\langle k \rangle = p (N-1)$, which will give you $p$.
>   * Now, use `networkx` to create the graph and extract the degree distribution.
>   * Finally, create a nice bar plot of the degree distribution, including axes labels and a plot title. Make sure that it looks like the Poisson distribution you plotted above.
> 

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import random

# generate 5000 numbers with a normal distribution 
# with mean 2 and standard deviation 0.125
numbers = np.random.normal(2, 0.125, 5000)
# create a histogram with 10 bins with numpy histogram function
hist, bins = np.histogram(numbers, bins=10)

bins =  0.5*(bins[1:] + bins[:-1])

plt.bar(bins, hist, width=bins[1]-bins[0], color='b')
plt.show()

plt.hist(numbers, bins=10, color='r',alpha=0.4)
plt.show()
```

```python
# 2nd task

random_ints = np.random.poisson(10,5000)
max_int = max(random_ints)
#print(max_int)
min_int = min(random_ints)
#print(min_int)

range_int = max_int - min_int
bins = np.arange(min_int, max_int, 1)
# np vector
hist, bins = np.histogram(random_ints, bins=bins)
bins =  0.5*(bins[1:] + bins[:-1])

plt.bar(bins, hist, width=1, color='b')
plt.show()
```

```python
# 3rd task

n = 5000
average_degree = 10
p = average_degree/(n-1)
erG = nx.erdos_renyi_graph(n, p)
#nx.draw(erG, with_labels=True, font_weight='bold')

degree_sequence = [d for n, d in erG.degree()]
min_degree_seq = min(degree_sequence)
max_degree_seq = max(degree_sequence)
range_int = max_int - min_int
bins = np.arange(min_int, max_int, 1)

# plot the degree distribution
hist, bins = np.histogram(degree_sequence, bins=bins)
bins =  0.5*(bins[1:] + bins[:-1])

plt.bar(bins, hist, width=bins[1]-bins[0], color='b')
# title, x and y labels
plt.title("Degree Histogram")
plt.ylabel("Count")
plt.xlabel("Degree")
plt.show()
```
